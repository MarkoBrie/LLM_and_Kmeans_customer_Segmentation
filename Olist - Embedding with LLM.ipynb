```python
import pandas as pd # dataframe manipulation
import numpy as np # linear algebra
from sentence_transformers import SentenceTransformer
```

```python
df = pd.read_csv("./kaggle/RFM_merged_15.csv")
```

```python
df.columns
```

```python
def compile_text(x):
    text =  f"""Customer city: {x['customer_city']},  
                Order status: {x['order_status']}, 
                Payment_type: {x['payment_type']}, 
                Product category: {x['product_category_name_english']}, 
                Review score: {x['review_score']}, 
                Recency: {x['Recency_y']}, 
                Monetary: {x['Monetary_y']}, 
                Frequency: {x['Frequency']}
            """

    return text
```

```python
def compile_text(x):
    text =  f"""Customer city: {x['customer_city']},
                Product category: {x['product_category_name_english']},
                Recency: {x['Recency_y']}, 
                Monetary: {x['Monetary_y']}, 
                Frequency: {x['Frequency']}
            """

    return text
```

```python
sentences = df.apply(lambda x: compile_text(x), axis=1).tolist()
```

```python
sentences
```

```python

model = SentenceTransformer(r"sentence-transformers/paraphrase-MiniLM-L6-v2")
```

```python

output = model.encode(sentences=sentences, show_progress_bar= True, normalize_embeddings  = True)
```

```python
df_embedding = pd.DataFrame(output)
df_embedding
```

```python




df_embedding.to_csv("embedding_train_Olist_15.csv",index = False)
```
